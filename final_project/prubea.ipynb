```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
class Polinomio (): 
    def __init__(self, C):
        self.coefs = C
    def evaluate(self, X): 
        coefs = self.coefs
        r = 0
        for c,i in zip(coefs, range(len(coefs))) : 
            r += c * (X ** i) 
        return r
```

```python
coefs1 = np.array([-1,2,3,-2.5])
coefs2 = np.array([1,0,-5,0,1])
P1 = Polinomio([-1,2,3,-2.5])
P2 = Polinomio([1,0,-5,0,1])

def coefs_dif(c1,c2) :
    n =  len(c1) - len(c2)
    if n == 0 :
        return c1 - c2
    elif n > 0 : 
        zeros_add = np.zeros(n)
        c2_new = np.append(c2, zeros_add)
        return c1 - c2_new
    else :
        zeros_add = np.zeros(-n)
        c1_new = np.append(c1, zeros_add)
        return c1_new - c2


A = np.flip(coefs_dif(coefs1,coefs2))

print(A)
np.roots(A)
```

```python
class Transform() : 
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def evaluate(self, x) : 
        T = (self.a + self.b ) / 2 + ((self.b - self.a ) / 2) * x
        return T
```
